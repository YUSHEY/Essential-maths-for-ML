```python
def f(x):
    return 2 * x + 1

x_values =[0,1,2,3]
for x in x_values:
    y=f(x)
    print(y)
    
```


 Functions are expressions that define relationships between two or more variables.
 More specifically, a function takes input variables (also called domain variables or
 independent variables), plugs them into an expression, and then results in an output
 variable (also called dependent variable)

```python
# charting a linear function
from sympy import * 
x=symbols('x')
f=2*x+1
plot(f)
```

#SYMPY
    SymPy is a Python library for symbolic mathematics. It provides tools for
    algebraic manipulation, calculus, equation solving, and more. SymPy allows you to
    work with mathematical expressions symbolically, meaning you can manipulate them
    algebraically rather than just numerically.

```python
from sympy import *
x=symbols('x')
f=x**2+1
plot(f)
```
